```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
import numpy as np
```

```python
df=pd.read_csv('../datasets/LogisticRegression/titanic.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")
imputer.fit(df.iloc[:, 5:6])
df.iloc[:, 5:6] = imputer.transform(df.iloc[:, 5:6])
df.iloc[:, 5:6]
df.isnull().sum()
```

```python
df=df.drop(['Embarked','Cabin','Ticket','PassengerId','Name','SibSp','Parch'],axis='columns')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Sex']=encoder.fit_transform(df['Sex'])
df.head()
```

```python
X=df.drop('Survived',axis="columns")
y=df['Survived']

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
# cross_val_score(LogisticRegression(),X,y,cv=10)
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model=SVC(max_iter=10000,tol=0.01,kernel="sigmoid")
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=RandomForestClassifier(n_estimators=50,criterion="entropy")
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=ExtraTreesClassifier(n_estimators=60)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=GaussianNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=MultinomialNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model=BernoulliNB()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
