```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../datasets/multiple_linear_regression_dataset.csv')
df.head()
```

```python
# plt.scatter(df['age'],df['income'],c="red",marker="+")
plt.scatter(df['experience'],df['income'],c="blue",marker="o")
```

```python
X=df.drop(['income'],axis="columns")
y=df['income']
X


model=LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
y_pred=model.predict(X)
y_pred[:5]
```

```python
model.predict([[43,20]])
```
